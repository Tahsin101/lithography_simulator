```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
def sqmat(msk):    
    shape=msk.shape[0] # shape of flattened array
    square_row=int(math.sqrt(shape))+1 # row of square matrix
    padwidth=abs(square_row**2-shape)
    msk=np.pad(msk,(0,padwidth),'edge')
    msk = msk.reshape((square_row,square_row)) # square matrix
    return msk
```

```python
df=pd.read_csv("image512_512.csv")
```

```python
I=df.iloc[:,0]
I=sqmat(I)
```

```python
plt.imshow(I,interpolation="gaussian")
plt.colorbar()
```
